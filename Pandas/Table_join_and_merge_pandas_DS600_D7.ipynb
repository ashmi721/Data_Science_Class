```python
# To read the employee table
emp = pd.read_excel('xlsx_data/employee-dept-table.xlsx', sheet_name='employee')
emp
```

```python
# To read the department table 
department = pd.read_excel('xlsx_data/employee-dept-table.xlsx', sheet_name= 'dept')
department
```

# merge employee and dept table

```python
# merge table with inner join(AnB)
pd.merge(left=emp, right=department, on='DepartmentID',how='inner')
```

* In inner join display the those data which is common to both table.
* In this condition we check the data on the basis of 
DepartmentID

```python
# merge table with outer join (AUB)
pd.merge(left=emp,right=department, on='DepartmentID', how='outer')
```

* In outer join display the hole data of the both table.

```python
# merge table with left join
pd.merge(left=emp,right=department, on='DepartmentID', how='left')
```

* In left join display only those data which are include only  A table or common of the both.

```python
# merge table with right join
pd.merge(left=emp,right=department, on='DepartmentID', how='right')
```

* In right join display only those data which are include only  b table or common of the both.

# categories of joins


```python
# one to one join
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
df3 = pd.merge(df1,df2)
df3
```
