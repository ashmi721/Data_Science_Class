```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv(r'C:\Users\ashmi\OneDrive\Desktop\Data Science\Data_Science_Class\Pandas\csv_data\covid_19_india.csv')
df.head()
```

```python
# Keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
# Reanaming columns name
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
df.tail(10)
```

```python

today = df[df.date=='2021-08-11']
today.head()
```

```python
# sorting data w.r.t number of confirmed cases
max_confirmed_cases = today.sort_values(by = 'confirmed',ascending=False)
max_confirmed_cases.head()
```

```python
# extract the top 5 most confirmed state
top_states_confirmed = max_confirmed_cases[0:5]
top_states_confirmed
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x = 'state',y='confirmed',data = top_states_confirmed,hue='state')
plt.title("Barchart of top 5 Confirmed State",fontsize=20)
plt.show()
```

* Maximum confirmed cases in maharastha

```python
# sorting data of death cases
max_death_cases = today.sort_values(by='deaths',ascending=False)
# Getting states with maximum number of death cases
top_states_death=max_death_cases[0:5]
top_states_death
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x = 'state',y='deaths',data = top_states_death,hue='state')
plt.title("Barchart of top 5 deaths State",fontsize=20)
plt.show()
```

This fig shows ,
* Maximum death state of covid-19 in this date 2021-08-11 is maharashtra(134201), Karnataka(36848) or etc.


```python
# sorting data of death cases
max_cured_cases = today.sort_values(by='cured',ascending=False)
# Getting states with maximum number of death cases
top_states_cured=max_cured_cases[0:5]
top_states_cured
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x = 'state',y='cured',data = top_states_cured,hue='state')
plt.title("Barchart of top 5 cured State",fontsize=20)
plt.show()
```
