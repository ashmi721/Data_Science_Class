# Line chart

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
growth = [3,3.5,5,5.9,7,9]
```

```python
! pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.plot() make the linechart
plt.plot(months,growth)
```

```python
plt.plot(months,growth,'o-') # o- is the line
```

```python
plt.plot(months,growth,'o--') # o-- is the dotline
```

```python
plt.plot(months,growth,'o-',color = 'purple') # change the color
```

```python
plt.plot(months,growth,'o-',color = 'purple')
plt.xlabel('Months')
plt.ylabel('Growth (%)')
plt.title('Monthly Growth in (%)')
```

```python
plt.plot(months,growth,'o-',color = 'purple')
plt.xlabel('Months')
plt.ylabel('Growth (%)')
plt.title('Monthly Growth in (%)')
plt.show() # it show only the plot
```

```python
plt.figure(figsize=(4,2)) # controls the size of a figure to be placed
plt.plot(months,growth,'o-',color = 'purple')
plt.xlabel('Months')
plt.ylabel('Growth (%)')
plt.title('Monthly Growth in (%)')
plt.show()
```

```python
plt.figure(figsize=(4,2)) # controls the size of a figure to be placed
plt.plot(months,growth,'o-',color = 'purple',ms=10)# ms means marker size o ko size ho
plt.xlabel('Months')
plt.ylabel('Growth (%)')
plt.title('Monthly Growth in (%)')
# plt.savefig('line.png') # saves the plot in png file
plt.show()
```

# Cricket data

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
actual_growth = [3,3.5,5,5.9,7,9]
predicted_growth=[2,2.7,4,5.5,7.6,8.5]
```

```python
plt.plot(months, actual_growth,'o-',label = 'Actual')
plt.plot(months,predicted_growth,'o-',label = 'Predicted')
plt.xlabel('Months')
plt.ylabel('Growth (%)')
plt.legend() # plt.legend() le plt.plot() ko label lai plot garxa
plt.show()
```

# Barplot

It is used to plot the categorycal values

```python
gender =['Male','Female']
frequency = [157,87]
```

```python
plt.bar(x = gender,height=frequency)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(x = gender,height=frequency , color = ['tab:blue','tab:orange']) # tab le dark color lai light banauxa
plt.xlabel('Gender')
plt.ylabel('Frequency/Count')
plt.show()
```

# Barplot using seaborn

```python
sns.barplot(x = gender,y = frequency) # matplotlib ma height hunxa seaborn ma teslai y lekhnu parxa
plt.xlabel('Gender')
plt.ylabel('Frequency/Count')
plt.show()
```

# Barplot using plotly

When we use plotly to make the Dynamic plot

```python
! pip install plotly
```

```python
import plotly.express as px
```

```python
fig = px.bar(x = gender,y = frequency,text=frequency) # text = frequency le bar ko top ma value dekhauxa
fig.update_layout(height=300,width = 400)
```

```python
fig = px.bar(x = gender,y = frequency,text=frequency) # text = frequency le bar ko top ma value dekhauxa
fig.update_layout(height=300,width = 400,xaxis_title='Gender',yaxis_title='Frequency/Count')
```

# piechart

```python
gender =['Male','Female']
frequency = [157,87]
```

```python
plt.pie(x=frequency,labels = gender, autopct = '%.2f%%')
plt.show()
```

```python
# using plotly
px.pie(names = gender,values=frequency)
```
