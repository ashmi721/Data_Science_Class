```python
data =df.groupby(df['species']).bill_length_mm.mean()
data
```

```python
fig, axes = plt.subplots()
sns.barplot(x=data.index,y=data.values,ax=axes)
axes.bar_label(axes.containers[0])
axes.set_title('Average of each species wise bill_length_mm ')
```

```python
fig,axes = plt.subplots()
axes.bar(x=data.index,height=data.values)
axes.bar_label(axes.containers[0])
axes.set_title('Average of each species wise bill_length_mm ')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
ax = axes[0, 0]
ax.bar(data.index, height=data.values,label='species')
# If you want to add labels to the bars, you can use bar_label method
ax.bar_label(ax.containers[0])
ax.legend()
plt.show()
```
