# Histogram, Kdeplot and Boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# load the iris dataframe
df = sns.load_dataset('iris')
df.head()
```

* Histogram

```python
plt.hist(df['sepal_length'])
```

```python
plt.hist(df['sepal_length'],bins=5) # We can control no.of bins in histogram
```

```python
# In seaborn and seaborn plot the best histogram ever matplotlib
sns.histplot(df['sepal_length'])
```

```python
sns.histplot(df['sepal_length'],kde=True)
```

```python
sns.kdeplot(df['sepal_length'])
```

# Boxplot

```python
plt.boxplot(df['sepal_width'])
```

```python
sns.boxplot(df['sepal_width'])
```

```python
# If you want to plot multiple data in boxplot to use seaborn
sns.boxplot(df)
```

```python
sns.boxplot(df)
sns.stripplot(df) # it shows the data point
```

```python
px.box(df.iloc[:,:-1])
```

```python
px.box(df.iloc[:,:-1],points='all')
```
