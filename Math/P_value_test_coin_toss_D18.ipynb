```python
H0 = 'The coin is unbiased'
H1 = 'The coin is biased'
```

```python
alpha = 0.05 # unless stated
```

```python
# say maile 10 times coin toss garey
n = 10
```

```python
no_of_heads = range(0,n+1)
```

```python
from scipy.special import comb
prob = []
p = 0.5
q = 1- p
for x in no_of_heads:
  px = comb(N=n, k=x)* p ** x * q ** (n-x)
  prob.append(px)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(20,8))
plt.bar(x= no_of_heads,height = prob)
ax.bar_label(ax.containers[0])
ax.set_xticks(no_of_heads)
fig.show()
```

```python
# p value for 9H 1 T
from scipy.stats import binomtest
result = binomtest( k=9, n= 10)
result
```

```python
result.pvalue
```

```python
if result.pvalue < alpha:
  print("We reject H0")
else:
  print('We fail to reject H0')
```
