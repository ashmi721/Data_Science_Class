# Kurtosis

Kurtosis refers to the degree of presence of outliers in the distribution. Kurtosis is a statistical measure, whether the data is heavy-tailed or light-tailed in a normal distribution.

Types of Kurtosis
 * MesoKurtic
 * platyKutic
 * LeptoKurtic

# MesoKurtic

* Normal distribution
* pearson Kurtosis nearly equals to 3
* Medium Outliers

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.normal(loc=5,scale=1,size=1500)

from scipy.stats import kurtosis
kurt = kurtosis(data,fisher=False) # use pearson method by disabling fisher

plt.figure(figsize=(4,2))
plt.title("Kurtosis={}".format(np.round(kurt,2)))
sns.histplot(data,kde=True)
plt.show()
```

# PlatyKurtic

* Uniforn distribution
* Kutosis < 3
* Low Outliers

```python
np.random.seed(0)
data = np.random.uniform(low=1,high=5,size = 1500)
from scipy.stats import kurtosis
kurt = kurtosis(data,fisher=False) # use pearson method by disabling fisher

plt.figure(figsize=(4,2))
plt.title("Kurtosis={}".format(np.round(kurt,2)))
sns.histplot(data,kde=True)
plt.show()
```

# LeptoKutic

* laplace distribution
* Kurtosis > 3
* High frequency of Outlier

```python
np.random.seed(0)
data = np.random.laplace(loc=5,scale=1,size=1500)

from scipy.stats import kurtosis
kurt = kurtosis(data,fisher=False) # use pearson method by disabling fisher

plt.figure(figsize=(4,2))
plt.title("Kurtosis={}".format(np.round(kurt,2)))
sns.histplot(data,kde=True)
plt.show()
```
